```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    
    @classmethod
    def pirate(cls, partner=''):
        aki = cls('Aki', 20)
        if partner:
            print(f'{aki.name} is running with {partner}.')
        else:
            print(f'{aki.name} is running.')
```
